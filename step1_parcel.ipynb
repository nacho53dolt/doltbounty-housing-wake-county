```python
import pandas as pd
import json, glob

with open(glob.glob('*parcel*.geojson')[0]) as fp:
    txt = fp.read()
    txt = '[' + txt.strip().replace('\n', ',\n') + ']'
    data = json.loads(txt)

for i in range(len(data)):
    data[i].update(data[i]['properties'])
```

```python

df = pd.DataFrame(data)

#df['latitude'] = df.geometry.apply(lambda x: x['coordinates'][1])
#df['longitude'] = df.geometry.apply(lambda x: x['coordinates'][0])
df = df.drop(columns=['properties', 'geometry', 'hash'])

df.to_csv('nc_wake_parcels.csv', index=False)
df
```
